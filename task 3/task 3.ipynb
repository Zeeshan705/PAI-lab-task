Solve the Water Jug Problem using Depth-First Search (DFS)


```python
MAX_A = 4
MAX_B = 3
GOAL = 2

visited = set()
path = []

def next_states(a, b):
    states = []

    states.append((MAX_A, b))   
    states.append((a, MAX_B))   
    states.append((0, b))       
    states.append((a, 0))       

    # pour A -> B
    t = min(a, MAX_B - b)
    states.append((a - t, b + t))

    # pour B -> A
    t = min(b, MAX_A - a)
    states.append((a + t, b - t))

    return states


def dfs(state):
    if state in visited:
        return False

    visited.add(state)
    path.append(state)

    a, b = state
    if a == GOAL or b == GOAL:
        return True

    for s in next_states(a, b):
        if dfs(s):
            return True

    path.pop()
    return False


start = (0, 0)

if dfs(start):
    print("Path:")
    for i in path:
        print(i)
else:
    print("No solution")
```
