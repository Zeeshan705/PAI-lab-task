Implement the N-Queens Problem (Dynamic)

```python
def print_board(board):
    for row in board:
        print(" ".join("Q" if x == 1 else "." for x in row))
    print()


def safe(board, row, col):
    n = len(board)

    for i in range(row):
        if board[i][col] == 1:
            return False
        if col-(row-i) >= 0 and board[i][col-(row-i)] == 1:
            return False
        if col+(row-i) < n and board[i][col+(row-i)] == 1:
            return False

    return True


def solve(board, row=0):
    n = len(board)

    if row == n:
        print_board(board)
        return True

    found = False
    for col in range(n):
        if safe(board, row, col):
            board[row][col] = 1
            found = solve(board, row+1) or found
            board[row][col] = 0

    return found


n = 4
board = [[0]*n for _ in range(n)]
print(solve(board))
```
